```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
X, y  = datasets.load_diabetes(return_X_y = True)#veri kumesini yükle
#true yazması x in bir veri matrisi olacagını y nin ise hedef değişken olacağını belirtir
X #özellikler (features)
y #hedef değişken (target)
print(X.shape)# veri kümesinin şekli
print(X[0])# veri kümesinin ilk satırı
```

```python
#Bu iki satır veri matrisini modellemeye uygun hale getirmek için yazılmış.
X = X[:, 3]#sadece 3.sütunu al veri setindeki tek bir ozelliği seciyoruz
X = X.reshape(-1,1)# tek sütunlu matris yap(-1 satır sayısını otomatik al 1 sütun sayısı)
#bunun amacı : sklearn in beklediği formatta olması
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()#scikit-learn’den bir Lineer Regresyon modeli oluşturuyor
model.fit(X_train, y_train)#modeli egitiyoruz
```

```python
y_pred = model.predict(X_test)
#En küçük kareler yöntemiyle 
# tahmin hatasını en aza indirecek doğruyu (doğru çizgiyi) bulur.
```

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
